```python
import matplotlib.pyplot as plt # Matikkakirjasto
import pandas as pd # IO-kirjasto
import numpy as np # Tieteellisen laskennan kirjasto

data = pd.read_excel('Gradia_JKL.xlsx') # Luetaan excel-file
data.head() # Tulostetaan headerit

```

```python
pvm = data['Date & Time'] # Valitaan sarakkeet omiksi muuttujikseen
lämpötila = data['Temp (degC)']
kosteus = data['Hum (%)']
tuulennopeus = data['Avg Wind Speed (m/s)']
```

```python
# Luodaan uusi kuvaaja
plt.figure()

# Tuodaan AutoDateLocator-työkalu matplotlibin dates-paketista
from matplotlib.dates import AutoDateLocator

# Asetetaan x-akselille automaattinen päivämääräasettelu
xtick_locator = AutoDateLocator()
ax = plt.gca()
ax.xaxis.set_major_locator(xtick_locator)

# Käännetään x-akselin otsikoita
plt.xticks(rotation=60)


plt.xlabel('pvm') # x-akselin otsikko
plt.ylabel('Lämpötila (°C)') # y-akselin otsikko
plt.title('Gradia JKL 3 vrk sää 12.4.2021') # Kuvaajan otsikko
plt.plot(pvm, lämpötila, label='Gradia JKL') # Mitä label tekee..? Tulostaa Labelin selitteeksi (legend)
plt.show() # Turha kun plt.plot() piirtää kuvaajan ilmankin?
```

```python
x = np.linspace(-10, 10, 50) # numpyn toiminto, luo numerot 0...10, jako 50:een 
y = -x**3-6*x**2 - x + 10 # Huom. eksponentti on **
# y = np.sin(x) # numpyn sin-funktio 
plt.plot(x, y, 'd', color='green') # 's' piirtää neliön, 'd' piirtää salmiakin
```

```python
x1 = np.linspace(0, 100, 30) # numpyn toiminto, luo numerot 0...10, jako 30:een 
y1 = np.exp(x1) # numpyn exp-funktio
plt.yscale('log') # Muutetaan plotin y-aks. 10-kant. logaritmiseksi
plt.plot(x1, y1, 's', color='orange') # 's' piirtää neliön, 'd' piirtää salmiakin
```
